# K-Means

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# load data from "kmeans.csv"
data = np.loadtxt("kmeans.csv")

# initialize k-means model
kmeans = KMeans(n_clusters=3)

# fit the model to the data
kmeans.fit(data)

# predict the cluster labels for each point
labels = kmeans.predict(data)

# plot the data points with their cluster labels
plt.scatter(data[:,0], data[:,1], c=labels)
plt.show()
```

# Mean shift

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import MeanShift

# load data from "kmeans.csv"
data = np.loadtxt("kmeans.csv")

# initialize mean shift model
ms = MeanShift()

# fit the model to the data
ms.fit(data)

# predict the cluster labels for each point
labels = ms.predict(data)

# plot the data points with their cluster labels
plt.scatter(data[:,0], data[:,1], c=labels)
plt.show()
```

```python
import numpy as np
from scipy.cluster.hierarchy import linkage, dendrogram
import matplotlib.pyplot as plt

# load data from data.txt
data = np.loadtxt("kmeans.csv")

# perform hierarchical clustering using linkage method 'ward'
Z = linkage(data, method='ward')

# plot dendrogram
dendrogram(Z)
plt.show()
```
